```
import pandas as pd
dataset = pd.read_csv('/content/drive/My Drive/Datasci/train.csv')
```

```
print(dataset)
```

```
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.catplot('OverallQual','SalePrice',data=dataset,hue='CentralAir',kind="point",height=11)
g.fig.suptitle('Overall Quality house and Air condition effect Houseprice')
g.set_xlabels('Overall Quality of the house')
g._legend.set_title("Air Condition")
print(g._legend)
```
